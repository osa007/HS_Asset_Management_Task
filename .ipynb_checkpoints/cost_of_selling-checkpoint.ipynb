```python
#function to get percentage of missing value
def missing_val(df):
    miss_val = df.isna().sum()
    miss_val_per = df.isna().sum()/len(df)*100
    miss_val_table = pd.concat([miss_val, miss_val_per], axis=1)
    miss_val_table_ren_columns = miss_val_table.rename(columns= {0 : 'MIssings Values', 1: '% of Total Value'})
    miss_val_table_ren_columns = miss_val_table_ren_columns[miss_val_table_ren_columns.iloc[:,:] !=0].sort_values(
    '% of Total Value', ascending=False).round(1)
    return miss_val_table_ren_columns
```

```python
# feature selection
def select_features(X, y, score_funcs, k='all'):
    # configure to select all features
    #f_regression or 
    fs = SelectKBest(score_func=score_funcs, k=k)
    # learn relationship from training data
    fs.fit(X, y)
    # transform train input data
    X_fs = fs.transform(X)
    return X_fs, fs
```

```python
# feature selection
X_fs, fs = select_features(X, y, score_funcs=mutual_info_regression, k='all')
# what are scores for the features
for i in range(len(fs.scores_)):
    print('Feature %d: %f' % (i, fs.scores_[i]))
# plot the scores
plt.bar([i for i in range(len(fs.scores_))], fs.scores_)
plt.show()
```

```python
cv = RepeatedKFold(n_splits=10, n_repeats=3, random_state=1)
# define the pipeline to evaluate
model = LinearRegression()
fs = SelectKBest(score_func=mutual_info_regression)
pipeline = Pipeline(steps=[('sel',fs), ('lr', model)])
# define the grid
grid = dict()
grid['sel__k'] = [i for i in range(X.shape[1]-20, X.shape[1]+1)]
# define the grid search
search = GridSearchCV(pipeline, grid, scoring='neg_mean_squared_error', n_jobs=-1, cv=cv)
# perform the search
results = search.fit(X, y)
# summarize best
print('Best MAE: %.3f' % results.best_score_)
print('Best Config: %s' % results.best_params_)
# summarize all
means = results.cv_results_['mean_test_score']
params = results.cv_results_['params']
for mean, param in zip(means, params):
    print(">%.3f with: %r" % (mean, param))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

    return(vif)

calc_vif(X)
```

#### Grid Search
For model developement stage, taking a statistical approach to determining the right algorithm to use for the model build 
framework is important

```python
%%time
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X1_train = sc_X.fit_transform(X_train)
X1_test = sc_X.transform(X_test)
```

```python
seed=7
kfold=KFold(n_splits=10)
scoring="r2"
```

```python
models=[]
models.append(["LR",LinearRegression()])
models.append(["SVR",SVR(gamma="scale")])
models.append(["KNR",KNeighborsRegressor()])
models.append(["RF",RandomForestRegressor()])
models.append(["XGB",XGBRegressor()])
```

```python
param_grids=[]
LR_param_grid={}
param_grids.append(LR_param_grid)
svr_param_grid={}
svr_param_grid["kernel"]=["linear","rbf","sigmoid"]
svr_param_grid["degree"]=[1,2,3,4]
svr_param_grid["C"]=[1,3,5,8]
param_grids.append(svr_param_grid)
knr_param_grid={}
knr_param_grid["n_neighbors"]=[3,5,7]
knr_param_grid["weights"]=["uniform","distance"]
param_grids.append(knr_param_grid)
rf_param_grid={}
rf_param_grid["n_estimators"]=list(np.arange(100,1000,100)) # start stop step
rf_param_grid["max_depth"]=[None,4,6,8,10,12,15]
rf_param_grid["max_features"]=[.2,.5,.8,1.0]
rf_param_grid["bootstrap"]=[True]
rf_param_grid["random_state"]=[seed]
param_grids.append(rf_param_grid)
xgb_param_grid={}
xgb_param_grid["n_estimators"]=list(np.arange(100,1000,100)) # start stop step
xgb_param_grid["max_depth"]=[None,4,6,8,10,12,15]
xgb_param_grid["max_features"]=[.2,.5,.8,1.0]
xgb_param_grid["random_state"]=[seed]
param_grids.append(xgb_param_grid)
```

```python
%%time
results=[]
for model,params in zip(models,param_grids): 
    grid_search=GridSearchCV(estimator=model[1],param_grid=params,cv=kfold,scoring=scoring) 
    grid_search.fit(X1_train,y_train) 
    results.append([model[0],grid_search.best_params_,grid_search.best_score_]) 
```

```python
print(results)
```

```python
%%time
#Check how model performed on the test set
test_scores=[]
for model,result in zip(models,results):
    regressor=model[1]
    regressor.set_params(**result[1])
    regressor.fit(X1_train,y_train)
    score=regressor.score(X1_test,y_test)
    test_scores.append([model[0],result[1],score])
```

```python
for model,param,score in test_scores:
    print("%s : %0.4f"%(model,score))
```

```python
#Funtion to calculate performance/evaluation metrics
def eval_metric(y_test, y_pred):  
    score = metrics.r2_score(y_test, y_pred)
    variance = metrics.explained_variance_score(y_test, y_pred)
    mse = metrics.mean_squared_error(y_test, y_pred)
    mae = metrics.mean_absolute_error(y_test, y_pred)
    print(f"The coefficient of determination score for the predicted cost of transaction is {score}")
    print(f"The explained varience score for the predicted cost of transaction is {variance}")
    print(f"The mean square error for the predicted cost of transaction is {mse}")
    print(f"The mean absolute error for the predicted cost of transaction is {mae}")
```
