```python
### TASK 2 Modelling 
```

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
%%time
#Splitting the dataset for train, test and eval. We would use 80% to train and split the remaining 20% (65:35)
X_train , X_test , y_train , y_test = train_test_split(X , y , 
                                            test_size = 0.15 ,
                                            random_state = 7)
```

```python
print('train data shape {} , test data shape {}'.format(X_train.shape,
                                                        X_test.shape))
```

```python
X_train
```

```python
%%time
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

```python
X_train
```

```python
seed=7
kfold=KFold(n_splits=10,random_state=seed)
scoring="r2"
```

Define model parameers for grid search

```python
models=[]
models.append(["LR",LinearRegression()])
models.append(["SVR",SVR(gamma="scale")])
models.append(["KNR",KNeighborsRegressor()])
models.append(["RF",RandomForestRegressor()])
```

```python
param_grids=[]
LR_param_grid={}
param_grids.append(LR_param_grid)
svr_param_grid={}
svr_param_grid["kernel"]=["poly","linear","rbf","sigmoid"]
svr_param_grid["degree"]=[1,2,3,4]
svr_param_grid["C"]=[.001,0.1,.3,.5,1,2,3]
param_grids.append(svr_param_grid)
knr_param_grid={}
knr_param_grid["n_neighbors"]=[3,5,7,11]
knr_param_grid["weights"]=["uniform","distance"]
param_grids.append(knr_param_grid)
rf_param_grid={}
rf_param_grid["n_estimators"]=list(np.arange(1,100,5)) # start stop step
rf_param_grid["max_depth"]=[None,2,3,4,5,6,7,8,9,10,12,15,18,20]
rf_param_grid["max_features"]=[.2,.3,.4,.5.6,.8,1.0]
rf_param_grid["bootstrap"]=[True]
rf_param_grid["random_state"]=[seed]
param_grids.append(rf_param_grid)
```

```python
%%time
results=[]
for model,params in zip(models,param_grids): 
    grid_search=GridSearchCV(estimator=model[1],param_grid=params,cv=kfold,scoring=scoring, n_jobs = -1) 
    grid_search.fit(X_train,y_train) 
    results.append([model[0],grid_search.best_params_,grid_search.best_score_]) 
```

```python
print(results)
```

```python
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
```

```python
print(best_accuracy)
```

```python
print(best_parameters)
```

```python
%%time
#Check how model performed on the test set
test_scores=[]
for model,result in zip(models,results):
    regressor=model[1]
    regressor.set_params(**result[1])
    regressor.fit(X_train,y_train)
    score=regressor.score(X_test,y_test)
    test_scores.append([model[0],result[1],score])
```

```python
for model,param,score in test_scores:
    print("%s : %0.4f"%(model,score))
```

```python
#Visualizing the model evaluation performance scores
evaluated_models=['LR', 'SVR', 'KNR', 'RForest']
evaluated_test_scores= [0.7056, 0.7454, 0.8084, 0.8847] 

plt.figure(figsize=(8,6))
plt.plot(evaluated_models,evaluated_test_scores,marker="o",linestyle="--",color="r")
plt.ylim(-0.5,1)
plt.title("Model Evaluation Plot")
plt.xlabel("Models")
plt.ylabel("R-Score")
plt.show()
```

```python
#Funtion to calculate performance/evaluation metrics
def eval_metric(y_test, y_pred):  
    score = metrics.r2_score(y_test, y_pred)
    variance = metrics.explained_variance_score(y_test, y_pred)
    mse = metrics.mean_squared_error(y_test, y_pred)
    mae = metrics.mean_absolute_error(y_test, y_pred)
    print(f"The house price model coefficient of determination score is {score}")
    print(f"The house price model explained varience score is {variance}")
    print(f"The house price model mean square error is {mse}")
    print(f"The house price model mean absolute error is {mae}")
```

MAE - meaning that larger errors will contribute linearly to the overall error. Like we’ve said above, a small MAE suggests the model is great at prediction, while a large MAE suggests that your model may have trouble in certain areas. A MAE of 0 means that your model is a perfect predictor of the outputs (but this will almost never happen).

While each residual in MAE contributes proportionally to the total error, the error grows quadratically in MSE. This ultimately means that outliers in our data will contribute to much higher total error in the MSE than they would the MAE
